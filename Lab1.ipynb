### Α) Δημιουργία και οπτικοποίηση σύνθετων τύπων δικτύου

Κατασκευή και οπτικοποίηση της τοπολογίας REG για βαθμό πλέγματος $d = 4$

```python
d = 4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. Επιλέξτε προσεκτικά τις παραμέτρους. 
#Θετουμε το p=0
REG =  nx.watts_strogatz_graph(n, d, 0, seed=None)
# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG,node_size=10,node_color="black",edge_color="red",width=0.1)
```

> Ας δούμε το αποτέλεσμα για ένα μικρότερο n για να γίνει ξεκάθαρη η μορφή του 
REG

```python
REG_more_clear =  nx.watts_strogatz_graph(10, 4, 0, seed=None)
# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG_more_clear,node_size=10,node_color="black",edge_color="red")
```

>Έτσι βλέπουμε ότι κατασκευάσαμε 2 regular (lattice) graph. Η κανονικότητα του εντοπίζεται στο γεγονός ότι κάθε κόμβος έχει βαθμό κ, άρα είναι κ-regular graph, και εδώ συγκεκριμένα είναι 4-κανονικός.<br/>
Στον δεύτερο γράφο που είναι πιο ξεκάθαρος μπορούμε να δούμε ότι ο κάθε κόμβος έχει ακμές με τους γειτονικούς του και με τους γείτονες των γειτόνων του, αλλά όχι self-loop με τον ευατό του.

Κατασκευή και οπτικοποίηση της τοπολογίας RG(ER) για αριθμό συνδέσεων $M = 750$

```python
n=190
M = 150
# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n, M)
# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER,node_size=10,node_color="black",edge_color="red",width=0.1)
```

Στους τυχαίους γράφους οι κόμβοι ενώνονται με ακμαίες που επιλέγονται τυχαία με ομοιόμορφο τρόπο.
Γενικά, οι τυχαίοι γράφοι μπορεί να κατακευαστούν με δύο **ισοδύναμους** τρόπους:
*   Gilbert G(n,p)
*   Erdos-Renyi G(N,M)

Εμείς χρησιμοποιήσαμε το μοντέλο Erdos-Renyi στο οποίο έχουμε έναν χώρο με όλους τους δυνατούς γράφους με n nodes και Μ ακμές και επιλέγουμε τυχαία με ομοιόμορφο τρόπο έναν από αυτούς.


Κατασκευή και οπτικοποίηση της τοπολογίας RGG σε περιοχή $L\times L = 1$ (unit cube) και ακτίνα σύνδεσης $R=0.25$

```python
R = 0.25
L=1
# για την κατασκευή του RGG, χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n, R) 
# για την οπτικοποίηση, χρησιμοποιήστε draw_network_edges, draw_network_nodes. 
# Για τη θέση των κόμβων στην τετράγωνη περιοχή, χρησιμοποιήστε την get_node_attributes από το networkx.
pos = nx.get_node_attributes(RGG, 'pos')
nx.draw_networkx_nodes(RGG, pos, node_size=10, node_color="black")
nx.draw_networkx_edges(RGG, pos, width=0.1, edge_color="red");
```

Οι Random Geometric Graphs είναι τυχαίοι χωρικοί γράφοι που επιτρέπουν την δημιουγεία ακμής αν βρίσκεται εντός μιας προκαθορισμένης πειροχής. Η τελική επιλογή της ακμής γίνεται πάλι με ομοιόμορφα τυχαίο τρόπο.

Κατασκευή και οπτικοποίηση της τοπολογίας SF για βαθμό αρχικού πλέγματος $d=4$

```python
d = 4
# για την κατασκευή του SF, χρησιμοποιήστε την barabasi_albert_graph από το networkx.
 
SF = nx.barabasi_albert_graph(n, d)
# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx.
nx.draw_circular(SF,node_size=10,node_color="black",edge_color="red",width=0.1)
```

Τα Scale-free (Exponential) Networks έχουν τα εξέις χαρακτηριστικά:

*   Ένα μικρό ποσοστό των κόμβων έχουν μεγάλο βαθμό
*   Οι περισσότεροι κόμβοι έχουν μικρό βαθμό <br>

Αυτό δημιουγεί το ποιοτικό αίσθημα της πυκνής πειροχής ακμών στο δεξί μέρος της εικόνας του γράφου, όπου εκέι εντοπίζεται κομβος ή κόμβοι με πολλούς γείτονες, δηλαδή δημοφιλής κόμβοι με μεγάλο βαθμό.
Αυτοί οι κόμβοι έχουν ιδιαίτερη αξία για το δίκτυο και ανάλογα με την εφαρμογή πχ σε ένα τηλεπικοινωνιακό δίκτυο αυτοί οι κόμβοι πρέπει να κάνουν έντονους ελέγχους συμφώρησης και πρέπει να προσπαθήσουμε να μην "πέσουν".<br/>
Μία άλλη ενδιαφέρουσα εφαρμογή βρίσκεται στα social media όπου αυτοί οι κόμβοι-ανθρωποι έχουν μεγάλο κοινό και επομένως αν κάνουν στοχευμένες διαγημήσεις θα επευθυνθούν σε περισσότερο κόσμο.<br/>
Τέλος σε αυτά τα δίκτυα ισχύει ότι "ο πλούσιος γίνεται πλουσιότερος και ο φτωχός φτωχότερος" λόγο του power-law.



Κατασκευή και οπτικοποίηση της τοπολογίας SW για βαθμό αρχικού πλέγματος $d=4$ και πιθανότητας ανασύνδεσης $p=0.3$

```python
d = 4
p = 0.3
# για την κατασκευή του SW, χρησιμοποιήστε την watts_strogatz_graph από το networkx.
SW = nx.watts_strogatz_graph(n, d, p)
# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx.
nx.draw_circular(SW,node_size=10,node_color="black",edge_color="red",width=0.1)
```

Ας ξεκινήσουμε με έναν εμπειρικό ορισμό των small world γράφων: "Οι περισσότεροι κόμβοι δεν είναι γείτονες, αλλά οι περισσότεροι κόμβοι μπορούν να πάνε ο ένας στον άλλον με ένα μικρό αριθμό hops, δηλαδή μπορώ να πάω από τον κόμβο Α στον κόμβο Β ταξιδεύοντας πάνω σε λίγες ακμές, και αυτό ισχύει για τους πειρσσότερους κόμβους Α,Β του δικτύου.
Επομένως βλέπουμε ότι σε ένα small word δίκτυο όλοι μπορούν να επικοινωνίσουν με τους άλλους όχι απευθείας αλλά με λίγα hops.

Το μοντέλο κατακευείς τους έιναι το Watts-Strogatz model:
*   Ξεκινάει με ένα κανονικό γράφο
*   Κάνει τυχαίο rewire για κάθε κόμβο με πιθανότητα p μη επιτέποντας  self-connections και duplicate edges
*   Μπορεί να προστεθούν αυθέρετες long-range edges

Επομένως μπρούμε να δούμε ότι αν το p=0 παραμένουμε σε ένα REG και αν το p=1 καταλλήγουμε σε έναν τυχαίο γράφο.

Σχετικά με το REG:<br/>
Ο REG είναι ένας κ-κανονικός γράφος, και συγκεκριμένα σε αυτή την εφαρμογή είναι ενας 4-κανονικός γράφος.
Αυτό σημαίνει ότι κάθε κόμβος του έχει βαθμό 4, δηλαδή ενόνεται με 4 διαφοερικούς κόμβους, αφου δεν υπάρχουν self-loops και duplicate ακμές/
Επομένως αφού κάθε κόμβος έχει βαθμό ίσο με 4 τόττε η μέση τιμή είναι 4 και η διασπορά μηδέν<br/>

Σχετικά με το RGER:<br/>
Για την κατασκευή του RGER έχουμε έναν χώρο με όλους τους δυνατούς γράφους με n nodes και Μ ακμές και επιλέγουμε τυχαία με ομοιόμορφο τρόπο έναν από αυτούς. Έτσι, ξέρουμε ότι η μέση τιμή θα είναι ίση με:
$${2*M \over N} = 7.89 $$
Και λόγο της τυχαιότητας της διαδικασίας σύνδεσεις κόμβων υπάρχει διασπορά ή οποία όμως δεν είναι πολύ μεγάλη δεδομένου ότι η διαδικασία γίνεται ομοιόμορφα.

Σχετικά με το RGG:<br/>
Οι Random Geometric Graphs είναι τυχαίοι χωρικοί γράφοι που επιτρέπουν την δημιουγεία ακμής αν βρίσκεται εντός μιας προκαθορισμένης πειροχής. Η τελική επιλογή της ακμής γίνεται πάλι με ομοιόμορφα τυχαίο τρόπο. Επίσης, τα RGG έχεουν μεγάλο var λόγω του της χωρικής τυχαιότητας της κατασκευής τους.
Επίσης, η μέση τιμή βαθμού υπολογίζεται ως:
$$z = {π r^2 \over L^2}n = 37.3 $$
Εδώ η διασπορά θα είνια μεγαλύερη λόγω της χωρικής τυχαιότητας της διαδικασίας, δηλαδή οι γείτονες εντώς των πυκνών περιοχών έχουν μεγαλύτερη πιθανόητα να έχουν πολλούς γείτονες ενώ αυτοί εκτός έχουν μεγάλη πιθανότητα να έχουν λίγους γείτονες. Φυσικά και στο var εξαρτάται το R, και εδώ εχουμε επιλέξει ενα συγκριτκά μεγάλο R.

Σχετικά με το SF:<br/>
Σε αυτά τα δίκτυα ισχύει ότι"ο πλούσιος γίνεται πλουσιότερος και ο φτωχός φτωχότερος" λόγο του power-law. Έτσι, η κατανομή των κόμβων ακολούθει power-law distribution. Αυτό εξηγεί και την μεγάλη διασπορά που έχουμε που αποκλίνει πολύ σε σχέση με την μέση τιμή του βαθμού κόμβου. Έτσι έχουμε λίγους κόμβους μεγάλου βαθμού με πολλούς κόμβους μικρού βαθμού.

Σχετικά με το SW:<br/>
Σε αυτά τα δίκτυα ισχύει ότι "Οι περισσότεροι κόμβοι δεν είναι γείτονες, αλλά οι περισσότεροι κόμβοι μπορούν να πάνε ο ένας στον άλλον με ένα μικρό αριθμό hops, δηλαδή μπορώ να πάω από τον κόμβο Α στον κόμβο Β ταξιδεύοντας πάνω σε λίγες ακμές, και αυτό ισχύει για τους πειρσσότερους κόμβους Α,Β του δικτύου. Για ένα δίκτυο με βάρη στις συνδέσεις του, όμοια με τον πίνακα γειτνίασης, ορίζεται ο πίνακας βαρών, $W=[w_{ij}]$, όπου με $w_{ij}>0$ συμβολίζεται το βάρος στην ακμή που ενώνει τους κόμβους $i,j$. Για μη-κατευθυνόμενους γράφους ο πίνακας $W$ είναι συμμετρικός.

```python
# για τιμές στο [1,10]: α=1, b=10, weight=np.random.rand()*(b-a) + a
# για την προσθήκη βαρών στις ακμές, χρησιμοποήστε την add.edge του networkx

def weights_array(G):
  weights_array= np.zeros((n,n))
  for x in G.edges:
    weights_array[x[0]][x[1]] = np.random.rand()*(9) + 1
    weights_array[x[1]][x[0]] = weights_array[x[0]][x[1]]
  return weights_array

def add_weighted_edge(G):
  weights = weights_array(G)
  for x in G.edges:
    G.add_edge(x[0],x[1], weight = weights[x[0]][x[1]])  
  return G
```

```python
#δίνεται η συνάρτηση strength που έχει ως όρισμα την τοπολογία που παράξατε στο ερώτημα Α με τα βάρη στις ακμές:
#python 3 implementation

def strength(topology):
    weights_list=[]
    for i in range(0,len(topology.adj)):
        a=[]
        for j in range (0,len(topology.adj[i].items())):
            d=list(topology.adj[i].values())[j]
            c=list(d.values())[0]
            a.append(c)
        weights_list.append(a)
    node_weights = []
    for node in weights_list:
        node_weights.append(sum(node))
    return node_weights

REG_strength=strength(REG)
RGER_strength=strength(RGER)
RGG_strength=strength(RGG)
SF_strength=strength(SF)
SW_strength=strength(SW)
```

**Degree Centrality:** 

Γενικά για την Degree Centrality έχουμε ορίσει δύο τρόπους μέτρησης η πρώτη είναι ο απλός ο βαθμός του κόμβου δηλαδή ο εξής τύπος:
$$ C_d(u) = {deg(u) } $$
ενώ η δεύτερη είναι ο βαθμός του κόμβου δια το πλήθος των μέγιστων δυνατο πλήθος γειτόνων που μπορεί να έχει δηλαδή δια του (n-1) δηλαδή ο εξής τύπος:
$$ C_d(u) = {deg(u) \over n-1} $$
Ο λόγος που γίνεται αυτή η κανονικοποίηση είναι για να έχουμε ένα πιο αντικειμενικό κριτήριο με βασή το μέγεθος n, και η αξία αυτής της ιδέας φαίνεται στο παράδειγμα του αστέρα όπως είδαμε και στο μάθημα θεωρίας.
Φυσικά ανάλογα με την εφραμογή θα μπορούσε να γινόταν διαφορετική κανονικοποίηση πχ να διερούσα με το μέγιστο node deggre του δικτύου

Μετά μπορούμε να μελετήσουμε την κεντρικοποίηση. Αφόρου υπολογίσουμε το centrality μπορεί να θέλυμε να υπολογίσουμε το dispresion του στους κόμβους.
Αυτό μπορεί να γίνει με την κλασσική στατιστική μελέτη της διασποράς ή με χρήση του τύπου του Freemans, που δεν είναι τίποτα άλλο από μια γενικευμένη var:

$${\displaystyle C_{D}(G)={\frac {\sum _{i=1}^{|V|}[C_{D}(v*)-C_{D}(v_{i})]}{|V|^{2}-3|V|+2}}}$$ όπου το $C_{D}(v*)$ είναι το μεγαλύτερο degree centrality ενός κόμβου στο γράφο.

**Closeness Centrality:**

Ή αλλιως proximity centrality και δειχνει την εγκιτητα και σχετίζεται με την απόσταση των κόμβων. Η ιδέα να αναζητήσουμε μια νέα μετρική ξεκινάει από την παρατήρηση ενός δικτύο γραμής όπου το deg καθε μη ακραίου κόμβου είναι ίσο με 2 και επομένως δεν έχουμε κάποια ουσιαστική πληροφορία.
Έτσι η κεντικότητα κάθε κόμβου ορίζεται ώς:
$$C_p(x) = \frac{1}{\sum_y d(y,x)}$$
Το αθροισμα αυτό είναι ολες οι αποστάσεις των κόμβων y από τον συγκεκριμένο κόμβο που μελετάμε προς όλους τους άλλους. Προφανώς μιλάμε για shortest paths.
Αυτό και πάλι θα μπορούσαμε να το μελετόυσαμε σε μια ανηγμένη μορφή αντίσοιχα όταν η μετρική απόστασης είναι τα hopς, η ανηγμένη μορφή θα είναι η εξής:
$$C_p(x) = \frac{n-1}{\sum_y d(y,x)}$$
Αν η μετρική δεν ειναι τα hops τότε δεν πρέπει να βάλω το n-1 αλλά με την μεγιστη απόσταση όπως προσδιορίζεται με τον τρόπο που μετράω την απόσταση.
Επειδή εδώ δεν ορίζεται η μετρική απόστασης δεν θα χρησιμοποιήσουμε κανονικοποιημένο τύπο.

Ο τύπος αυτός λαμβάνει υπόψην όλο το εύρος αποστάσεων επομένως δεν είναι τοπικό κριτήριο αλλά από την άλλη ο υπολογισμός του είναι αρκετά πιο σύνθετος.

**Betweenness Centrality**

Θέωρωουμε ένα δικτυο και εξετάζουμε την ροή της πληροφορίας, με την έννια το ποιος ελέγχει την ροή της πληροφορίας. Πόσους πρέπει εγώ να περάσω για να μεταφέρω την πληρφορία μου από εδω μου είμαι εγώ για να πάω σε κάποιον άλλον, με άλλα λόγια από πόσους εξαρτάμαι για την μεταφρορα της πληροφρίας μου, ή πόσοι ελέγχουν την μεταφορά της πληροφορίας μου.
Η προηγουμενη μετρική μπορεί να μας δώσει την πληροφορία για το πόσο γρήγορα θα διατρέξω αυτό το μονοπάτι, αλλά τωρα θέλουμε να δούμε το κατα πόσο ένας κόμβος μπορεί να ελέγξει αυτή την πληροφορία.
Π.χ. Αυτό είναι πολύ σημαντικό σε τηλεπικοινωνιακά δίκτυα πχ αν χαλάσει ένα πολύ σημαντικό κομμάτι που ελέγχει πολλές δικτιακές διαδρομές τότε θα έχουμε σημαντικό πρόβλημα.
Για να το υπολογίσω υπολογίζω πρώτα το partial betweenness, αν ενας κόμβος $p_k$ είναι αποσυνδεδμένος τότε θεωρώ την τιμή του ίση με το 0, ορίζω ως $ g_ij$ το πλήθος των συντομοτερων μονοπατίων μεταξυ των i και j, από αυτά κάποια περνάνε από τον κόμβο $p_k$  και τα ονομάζω ως $ g_ij(p_k)$.

Έτσι partial betweeness του κόμβου $p_k$ είναι:
$$ b_ij(p_k) = {g_ij(p_k) \over g_ij } $$

- degree centrality
- closeness centrality
- betweenness centrality
- eigenvector centrality

eigenvector centrality: Relative scores are assigned to all nodes in the network based on the concept that connections to high-scoring nodes contribute more to the score of the node in question than equal connections to low-scoring nodes. A high eigenvector score means that a node is connected to many nodes who themselves have high scores.[

```python
def calculate_degree_centrality(G):
  degree_centrality = nx.degree_centrality(G)
  return degree_centrality

def calculate_closeness_centrality(G):
  closeness_centrality = nx.closeness_centrality(G)
  return closeness_centrality

def calculate_betweenness_centrality(G):
  betweenness_centrality = nx.betweenness_centrality(G)
  return betweenness_centrality

def calculate_eigenvector_centrality(G):
  eigenvector_centrality = nx.eigenvector_centrality(G,max_iter=666)
  # eskage stis default 100 epanalipseis opote bazw enan megalutero ari8mo
  return eigenvector_centrality

REG_degree_centrality = calculate_degree_centrality(REG)
REG_closeness_centrality = calculate_closeness_centrality(REG)
REG_betweenness_centrality = calculate_betweenness_centrality(REG)
REG_eigenvector_centrality = calculate_eigenvector_centrality(REG)

RGER_degree_centrality = calculate_degree_centrality(RGER)
RGER_closeness_centrality = calculate_closeness_centrality(RGER)
RGER_betweenness_centrality = calculate_betweenness_centrality(RGER)
RGER_eigenvector_centrality = calculate_eigenvector_centrality(RGER)

RGG_degree_centrality = calculate_degree_centrality(RGG)
RGG_closeness_centrality = calculate_closeness_centrality(RGG)
RGG_betweenness_centrality = calculate_betweenness_centrality(RGG)
RGG_eigenvector_centrality = calculate_eigenvector_centrality(RGG)

SF_degree_centrality = calculate_degree_centrality(SF)
SF_closeness_centrality = calculate_closeness_centrality(SF)
SF_betweenness_centrality = calculate_betweenness_centrality(SF)
SF_eigenvector_centrality = calculate_eigenvector_centrality(SF)

SW_degree_centrality = calculate_degree_centrality(SW)
SW_closeness_centrality = calculate_closeness_centrality(SW)
SW_betweenness_centrality = calculate_betweenness_centrality(SW)
SW_eigenvector_centrality = calculate_eigenvector_centrality(SW)

# για πιο εύκολο σχολιασμό, να παράξετε 1 figure για όλες τις τοπολογίες ανά κεντρικότητα
def plot_centrality(centrality, REG_centrality,RGER_centrality,RGG_centrality,SF_centrality,SW_centrality):
  plt.figure()
  name = centrality.split("_")
  plt.title("Cumulative "+name[0]+" "+name[1]+" distributions")
  plt.xlabel(name[0]+" "+name[1])
  plt.ylabel("distributions")
  plt.grid()
  arr1 = np.array( list(REG_centrality.values()))
  arr1.sort
  plt.plot(arr1,[x / n for x in list(range(n))],color="black")
  arr2 = np.array(list(RGER_centrality.values()))
  arr2.sort()
  plt.plot(arr2,[x / n for x in list(range(n))],color="red")
  arr3 = np.array(list(RGG_centrality.values()))
  arr3.sort()
  plt.plot(arr3,[x / n for x in list(range(n))],color="brown")
  arr4 = np.array(list(SF_centrality.values()))
  arr4.sort()
  plt.plot(arr4,[x / n for x in list(range(n))],color="blue")
  arr5 = np.array(list(SW_centrality.values()))
  arr5.sort()
  plt.plot(arr5,[x / n for x in list(range(n))],color="pink")
  for_legend =['REG','RGER','RGG','SF','SW']
  plt.legend(for_legend)
  return
```

Επομένως δεν παρουσιάζει κάποιο ενδιφέρον η ανάλυση του.

Για το RGER model και το RGG model, παρατηρούμε παρόμοια ποιοτικά αποτελέσματα ως προς τις κατανομές των centralities τους. Από την κατανομή της degree centrality, καταλαβαίνουμε πως προσομοιάζει αρκετά στην Uniform distribution, αφού παρατηρούμε και για τα δύο μοντέλα παρόμοιες αυξήσεις στις μετρήσεις της κεντρικότητας, στην συγκεντρωτική κατανομή, καθώς αυξάνεται ο αριθμός των κόμβων. Στις υπόλοιπες κεντρικότητες, παρατηρούμε παρόμοια συμπεριφορά και στα δυο μοντέλα, με την διαφορά ότι στο RGG model έχουμε μεγαλύτερη αύξηση κεντρικότητας από ότι στο RGER στους πρώτους κόμβους που μελετάμε. Στην συνέχεια, βλέπουμε ποιοτικά παρόμοια συμπεριφορά στην κατανομή των κεντρικοτήτων.

Για το SF model και το SW model, παρατηρούμε ότι οι κατανομές των κεντρικοτήτων τους ποικίλουν. Πιο αναλυτικά, βλέπουμε ότι για το Degree centrality, το πρώτο παρομοιάζει με Poisson distribution, ενώ το δεύτερο με Uniform distribution, για το closeness centrality και τα δύο παρομοιάζουν με παρόμοιες κατανομές, με το πρώτο να έχει την απότομη αύξηση πιο νωρίς από το δεύτερο. Επιπλέον, για την betweenness centrality παρατηρούμε ότι το SF παρομοιάζει με την power-law distribution, βλέποντας λίγους κόμβους με μεγάλη κεντρικότητα και πολλούς με μικρή, κάτι το οποίο δεν φαίνεται για το SW, το οποίο εμφανίζει μεγαλύτερη ομοιμορφία ως προς αυτή την κεντρικότητα. Τέλος, για την eigenvector centrality, τα δυο μοντέλα εμφανίζουν ομοιότητες, με το SF να εμφανίζεται πιο κοντά στην power-law distribution, ενώ το SW πιο κοντά στην heavy-tail distribution.

Σχολιασμός εξελικτικής μετατροπής:

Ας ξεκινήσουμε με έναν εμπειρικό ορισμό των small world γράφων: "Οι περισσότεροι κόμβοι δεν είναι γείτονες, αλλά οι περισσότεροι κόμβοι μπορούν να πάνε ο ένας στον άλλον με ένα μικρό αριθμό hops, δηλαδή μπορώ να πάω από τον κόμβο Α στον κόμβο Β ταξιδεύοντας πάνω σε λίγες ακμές, και αυτό ισχύει για τους πειρσσότερους κόμβους Α,Β του δικτύου.
Επομένως βλέπουμε ότι σε ένα small word δίκτυο όλοι μπορούν να επικοινωνίσουν με τους άλλους όχι απευθείας αλλά με λίγα hops.

Το μοντέλο κατακευείς τους έιναι το Watts-Strogatz model:
*   Ξεκινάει με ένα κανονικό γράφο
*   Κάνει τυχαίο rewire για κάθε κόμβο με πιθανότητα p μη επιτέποντας  self-connections και duplicate edges
*   Μπορεί να προστεθούν αυθέρετες long-range edges

Επομένως μπρούμε να δούμε ότι αν το p=0 παραμένουμε σε ένα REG και αν το p=1 καταλλήγουμε σε έναν τυχαίο γράφο.

Έτσι γενικά βλέπουμε ότι:


*   Για $p=0$ προκύπτει REG με τον ίδιο ΣΟ και mean path length που είχε υπολογιστεί σε προηγούμενο ερώτημα.
*   Για $p>0$ και $ p <1 $ έχουμε small word δίκτυα.
*   Για $p=1$ προκύπτει τυχαίος γράφος με τον ίδιο ΣΟ και mean path length που είχε υπολογιστεί σε προηγούμενο ερώτημα, καθώς έχουμε αναδιόρθωση ακμών με τυχαίο τρόπο.

### Κ) Εγω-κεντρικότητες (Ego-centralities)
Για τις συνθετικές τοπολογίες του πίνακα 1, να υπολογιστεί η ενδιαμεσική εγω-κεντρικότητα (ego-betweenness centrality) του κάθε κόμβου. Στη συνέχεια να καταταχθούν οι κόμβοι με βάση την ενδιαμεσική εγωκεντρικότητα και την ενδιαμεσική κεντρικότητα και να συγκριθούν/σχολιαστούν οι κατατάξεις ως προς την ομοιότητά τους για κάθε μία τοπολογία.  

```python
n=190
d=4
M=750
R=0.25
p=0.3
REG =  nx.watts_strogatz_graph(n, d, 0, seed=None)
RGER = nx.gnm_random_graph(n, M)
RGG = nx.random_geometric_graph(n, R) 
SF = nx.barabasi_albert_graph(n, d)
SW = nx.watts_strogatz_graph(n, d, p)
```

```python
#για την κατασκευή των εγωδικτύων, χρησιμοποιήστε την ego_graph του networkx
#για τον υπολογισμό της ενδιαμεσικής εγωκεντρικότητας θα χρειαστεί να δημιουργήσετε τον πίνακα γειτνίασης του εγωδικτύου
#Χρήσιμες συναρτήσεις: neighbors(networkx), ones(numpy),linalg.matrix_power, multiply, triu, nonzero του numpy
#κατασκευάστε dataframe από το pandas με γραμμές τους κόμβους και στήλες τις τιμές των υπό εξέταση κεντρικοτήτων
#κατασκευάστε στη συνέχεια pairplot από το seaborn για να εξετάσετε τις συσχετίσεις μεταξύ τους.

def keep_possitive_and_sum_inverse_digits(arr):
  sum =0
  for i,val1 in enumerate(arr):
    for j,val in enumerate(arr[i]):
      if(val<=0):
        continue
      sum+=1/val
  return sum

def Ego_centralities(G):
    ego_centrality_list = []
    for node in G.nodes:
        node_ego = nx.ego_graph(G, node, radius=1) #induced subgraph me nodes olous tous geitones tou node
        # adjacency matrix
        # to nx.adj_matrix(node_ego) einai scipy.sparse.csr.csr_matrix
        # to 8elw na to kanw omws <class 'numpy.ndarray'>
        A = (nx.adj_matrix(node_ego)).toarray()
        hadamard_mul = np.multiply(np.matmul(A,A), (1-A))  # kane A^2 * (1-A)
        anw_trigonikos_arr = np.triu(hadamard_mul, 1) #krataw anw trigwniko
        ego_betweenness= keep_possitive_and_sum_inverse_digits(anw_trigonikos_arr) # mono 8etika kai kane to Σ (1/val 
        ego_centrality_list.append(ego_betweenness)
    return ego_centrality_list



print("The following array is the ego-centralities for each node of the REG:")
REG_Ego_centralities = Ego_centralities(REG)
print(REG_Ego_centralities)
print()

print("The following array is the ego-centralities for each node of the RGER:")
RGER_Ego_centralities= Ego_centralities(RGER)
print(RGER_Ego_centralities)
print()

print("The following array is the ego-centralities for each node of the RGG:")
RGG_Ego_centralities = Ego_centralities(RGG)
print(RGG_Ego_centralities)
print()

print("The following array is the ego-centralities for each node of the SF:")
SF_Ego_centralities = Ego_centralities(SF)
print(SF_Ego_centralities)
print()

print("The following array is the ego-centralities for each node of the SW:")
SW_Ego_centralities = Ego_centralities(SW)
print(SW_Ego_centralities)
print()
```
